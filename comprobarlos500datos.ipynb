```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score

# Cargar el modelo
modelo = joblib.load("modelo_alzheimer.pkl")

# Cargar el archivo con los datos y el diagnóstico real
df = pd.read_excel("500 de comprobacion.xlsx")

# Separar X (features) y y (objetivo real)
X = df.drop(columns=["Diagnosis"])
y_real = df["Diagnosis"]

# Obtener predicciones del modelo
y_pred = modelo.predict(X)

# Comparar y mostrar caso por caso
correctos = 0
for i in range(len(y_real)):
    real = y_real.iloc[i]
    pred = y_pred[i]
    resultado = "✅ Correcto" if real == pred else "❌ Incorrecto"
    estado = "Tiene Alzheimer" if pred == 1 else "No tiene Alzheimer"
    print(f"Paciente {i+1}: Predicción → {estado} | Real → {'Tiene Alzheimer' if real == 1 else 'No tiene Alzheimer'} → {resultado}")
    if real == pred:
        correctos += 1

# Mostrar precisión total
accuracy = correctos / len(y_real) * 100
print(f"\n📊 Precisión total del modelo: {accuracy:.2f}%")
```
